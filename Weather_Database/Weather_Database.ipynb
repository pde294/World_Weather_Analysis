```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
new_coordinates = list(lat_lngs)
```

```python
from citipy import citipy
```

```python
# Create a list for holding the cities.
new_cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in new_coordinates:
    new_city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if new_city not in new_cities:
        new_cities.append(new_city)
# Print the city count to confirm sufficient count.
len(new_cities)
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Import the datetime module from the datetime library.
from datetime import datetime
```

```python
# Create an empty list to hold the weather data.
new_city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

# Loop through all the cities in the list.
for i, new_city in enumerate(new_cities):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    new_city_url = url + "&q=" + new_city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {new_city}")
    # Add 1 to the record count.
    record_count += 1# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        new_city_weather = requests.get(new_city_url).json()
        # Parse out the needed data.
        new_city_lat = new_city_weather["coord"]["lat"]
        new_city_lng = new_city_weather["coord"]["lon"]
        new_city_max_temp = new_city_weather["main"]["temp_max"]
        new_city_humidity = new_city_weather["main"]["humidity"]
        new_city_clouds = new_city_weather["clouds"]["all"]
        new_city_wind = new_city_weather["wind"]["speed"]
        new_city_country = new_city_weather["sys"]["country"]
        # Convert the date to ISO standard.
        new_city_date = datetime.utcfromtimestamp(new_city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        new_city_data.append({"City": new_city.title(),
                          "Lat": new_city_lat,
                          "Lng": new_city_lng,
                          "Max Temp": new_city_max_temp,
                          "Humidity": new_city_humidity,
                          "Cloudiness": new_city_clouds,
                          "Wind Speed": new_city_wind,
                          "Country": new_city_country,
                          "Date": new_city_date})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
    
    
```
